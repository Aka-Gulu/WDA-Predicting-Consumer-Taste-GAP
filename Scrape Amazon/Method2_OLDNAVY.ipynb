```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from webdriver_manager.microsoft import EdgeChromiumDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
import urllib.request, urllib.error, urllib.parse
import json
import time
import random
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Configure Edge WebDriver options
edge_options = Options()
edge_options.use_chromium = True  # Use Chromium-based Edge

# Set up Edge WebDriver service
edge_service = Service(EdgeChromiumDriverManager().install())

# Create an Edge WebDriver instance with the specified options
driver = webdriver.Edge(service=edge_service, options=edge_options)



brand_names=[]
product_names= []
product_ratings = []
review_count = []
product_price= []
url_list = ['https://www.amazon.com/s?k=trending+clothes&crid=B9J16A5B5HGZ&sprefix=trending+clothes%2Caps%2C117&ref=nb_sb_noss_1',
            'https://www.amazon.com/s?k=trending+clothes&page=2&crid=B9J16A5B5HGZ&qid=1696183064&sprefix=trending+clothes%2Caps%2C117&ref=sr_pg_1',           
            'https://www.amazon.com/s?k=trending+clothes&page=3&crid=B9J16A5B5HGZ&qid=1696183226&sprefix=trending+clothes%2Caps%2C117&ref=sr_pg_2',
            'https://www.amazon.com/s?k=trending+clothes&page=4&crid=B9J16A5B5HGZ&qid=1696183253&sprefix=trending+clothes%2Caps%2C117&ref=sr_pg_3',
            'https://www.amazon.com/s?k=trending+clothes&page=5&crid=B9J16A5B5HGZ&qid=1696183273&sprefix=trending+clothes%2Caps%2C117&ref=sr_pg_4']

for url in url_list:
    driver.get(url)

    time.sleep(5)

    html = driver.page_source 

    soup = bs(html,'html.parser')

    bra_name = soup.find_all('span', class_="a-size-base-plus a-color-base")
    for x in bra_name:
        name = x.text
        print(name)
        brand_names.append(name)
    
    pro_name = soup.find_all('span', class_="a-size-base-plus a-color-base a-text-normal")
    for x in pro_name:
        name = x.text
        print(name)
        product_names.append(name)
    
    pro_rate = soup.find_all('span', class_="a-size-base puis-normal-weight-text")
    for x in pro_rate:
        rate = x.text
        print(rate)
        product_ratings.append(rate)
    
    review_cou = soup.find_all('span', class_="a-size-base s-underline-text")
    for x in review_cou:
        review = x.text
        print(review)
        review_count.append(review)

    price = soup.find_all('span', class_="a-price-whole")
    for x in price:
        xx = x.text
        print(xx)
        product_price.append(xx)
```

```python
import csv

# Assuming you have the review_count list populated from your scraping code

# Remove non-breaking space characters and parentheses from the review_count list
review_count = [review.replace('\xa0', '').strip('()') for review in review_count]

output_file = "Files_directory/method_2_OLDNAVY.csv"
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Brand Name', 'Product Name', 'Product Rating', 'Review Count', 'Product Price'])  # Write header row

    # Write data rows
    for i in range(len(review_count)):
        csv_writer.writerow([brand_names[i], product_names[i], product_ratings[i], review_count[i], product_price[i]])
```
